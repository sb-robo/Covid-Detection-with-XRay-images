```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

### Import Scratch History

```python
hist_scratch = np.load('./Model_history/scratch_file.npy',allow_pickle='TRUE').item()
hist_vgg = np.load('./Model_history/vgg_file.npy',allow_pickle='TRUE').item()

print(hist_scratch.keys(),hist_vgg.keys())
```

```python
x = np.arange(1,11)
print(x)
```

```python
fig,ax = plt.subplots(1,2,figsize=(16,5))

ax[0].plot(x,hist_scratch['loss'],label='Training Loss')
ax[0].plot(x,hist_scratch['val_loss'],label='Validation Loss')
ax[0].set_xlabel('Epochs',fontsize=14)
ax[0].set_ylabel('Loss',fontsize=14)
ax[0].set_xlim(1,10)
ax[0].set_title('Scratch Model: Loss vs Epochs',fontsize=20)
ax[0].legend()

ax[1].plot(x,hist_vgg['loss'],label='Training Loss')
ax[1].plot(x,hist_vgg['val_loss'],label='Validation Loss')
ax[1].set_xlabel('Epochs',fontsize=14)
ax[1].set_ylabel('Loss',fontsize=14)
ax[1].set_xlim(1,10)
ax[1].set_title('VGG Model: Loss vs Epochs',fontsize=20)
ax[1].legend()

plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(16,5))

ax[0].plot(x,hist_scratch['accuracy'],label='Training Accuracy')
ax[0].plot(x,hist_scratch['val_accuracy'],label='Validation Accuracy')
ax[0].set_xlabel('Epochs',fontsize=14)
ax[0].set_ylabel('Accuracy',fontsize=14)
ax[0].set_xlim(1,10)
ax[0].set_title('Scratch Model: Accuracy vs Epochs',fontsize=20)
ax[0].legend()

ax[1].plot(x,hist_vgg['accuracy'],label='Training Accuracy')
ax[1].plot(x,hist_vgg['val_accuracy'],label='Validation Accuracy')
ax[1].set_xlabel('Epochs',fontsize=14)
ax[1].set_ylabel('Accuracy',fontsize=14)
ax[1].set_xlim(1,10)
ax[1].set_title('VGG Model: Accuracy vs Epochs',fontsize=20)
ax[1].legend()

plt.show()
```
