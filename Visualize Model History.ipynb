```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
```

### Import History

```python
hist_scratch = np.load('./Model_history/scratch_file.npy',allow_pickle='TRUE').item()

print(hist_scratch.keys())
```

```python
x = np.arange(1,11)
print(x)
```

```python
plt.plot(x,hist_scratch['loss'],label='Training Loss')
plt.plot(x,hist_scratch['val_loss'],label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xlim(1,10)
plt.title('Loss vs Epochs')
plt.legend()
plt.show()
```

```python
plt.plot(x,hist_scratch['accuracy'],label='Training Accuracy')
plt.plot(x,hist_scratch['val_accuracy'],label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.xlim(1,10)
plt.title('Accuracy vs Epochs')
plt.legend()
plt.show()
```
